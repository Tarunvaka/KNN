```python
df['diagnosis'] = df['diagnosis'].map({'M':1, 'B':0})
```

```python
X = df.drop(columns='diagnosis')
y = df['diagnosis']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
k = 5
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print("Classification Reports")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(x_scaled)
```

```python
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca, y,test_size=0.2)
```

```python
knn.fit(X_train_pca, y_train_pca)
```

```python
x_min, x_max = X_pca[:,0].min()-1, X_pca[:,0].max()+1
y_min, y_max = X_pca[:,1].min()-1, X_pca[:,1].max()+1
xx,yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                    np.arange(y_min, y_max, 0.1))
```

```python
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
```

```python
plt.figure(figsize=(10,6))
plt.contour(xx,yy,Z,alpha = 0.8, cmap='coolwarm')
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=df['diagnosis'], palette='coolwarm', edgecolor = 'k', s = 50)
plt.show()
```
